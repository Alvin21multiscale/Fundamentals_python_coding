# Some simple cleaning exercises for strings and texts

```python
# Disclaimer - The following is not the only way to perform some simple exercises for strings and texts in python

# stripping unwanted characters from strings using strip()
import re # import regular expression function in python
import random # import random function in python
import string # import string function in python 
letters = string.ascii_letters # range of upper- and lower-case letters in python
size = random.randint(1,10e4) # size of arbitrary string or text 

new_string = '' # define an empty string 
for i in range(value):
    dummy = random.randint(1,2) # computer randomly chose 1 or 2
    if dummy == 1:
        new_string += " " # if 1, add a whitespace
    elif dummy == 2:
        new_string += random.choice(letters) # else, add an arbitrary letter
        
result_1 = new_string.strip() # remove whitespaces from the beginning or end of new_string
result_2 = new_string.lstrip() # remove whitespaces from the left side of new_string
result_3 = new_string.rstrip() # remove whitespaces from the right side of new_string
result_4 = new_string.replace(' ', '') # remove all whitespaces with no spaces
result_5 = ((re.sub('\s+', ' ',new_string)).lstrip()).rstrip() # only remove selected whitespaces. 
                                                               # Note that \s+ refers to a string
print(result_1)
print('')
print(result_2)
print('')
print(result_3)
print('')
print(result_4)
print('')
print(result_5)
print('')

# searching and replacing text
modified_string = result_5.replace('D' or 'd', 'DOGGIE!')
modified_string = modified_string.replace('m' or 'M', '$$$$')
print(modified_string)
print('')

# simple cleaning of date string
date_1 = '24/12/2019'
clean_test = re.compile(r'(\d+)/(\d+)/(\d+)')
cleaned_date = clean_test.sub(r'\1-\2-\3', date_1)
print(cleaned_date)
```
